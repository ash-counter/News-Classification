# Regex
So you want to match patterns effectively in a string, to effectively clean the textual data, which needs to be processed. Effective way to do is through the use of *Regular*(***Reg***) *Expressions*(***Ex***) or ***Regex*** library. For this we just have to match patterns in a given text and format according to our needs. Generally it involves removing the special characters which mean less to NLP algorithms. 
```python
import re
example_text="anything you want to have"
pattern=re.compile(r'') #anything between single quotes directs the regex to compile the pattern we would look for in the text.
matched=pattern.finditer(example_text) #then you iterate over the text looking for patterns and storing any matched values.
```
1. Different shortcuts used with ```re.compile(r'')``` 
    1. ```r'.'``` = Any character except newline
    2. ```r'\d'```= digits [0-9]
    3. ```r'\w'```= word character [a-z],[A-Z],[0-9],[underscore]
    4. ```r'\s'```= whitespace(spaces,tabs,newline)
2. Anchors which are used to define spaces in the text
    1. ```r'\bsome_text'```= '\b' stands for Word boundary so it will search for some_text with the word boundary.
    2. ```r'^some_text'```='^' start of the string. Therefore this matches only when some_text is present at the start of string.
    3. ```r'some_text$'```='$' start of the string. Therefore this matches only when some_text is present at the start of string.
3. Brackets
    1. ```r'[]'```= matches character in the brackets
    2. ```r'[^]'```= matches character NOT in the brackets
    3. ```r'|'``` = Either or
    4. ```r'()'```= Group by
4. Quantifiers
    1. ```*```= 0 or more
    2. ```+```= 1 or more
    3. ```?```= 0 or 1
    4. ```{3}```= Exact number
    5. ```{3,6}```= Range of numbers
5. Methods associated with Regex
    1. finditer
    2. findall
    3. match
    4. search
6. Flags

````python
example_text="anything you want to have"
pattern= re.compile(r'.')
matched=pattern.finditer(example_text)
for match in matched:
    print(match)
````
